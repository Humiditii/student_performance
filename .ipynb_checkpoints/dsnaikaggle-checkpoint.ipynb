```python
def operate_empty(data):
    for x in data.columns:
        y = (data[x].isnull().sum() / len(data)) * 100 
        threshold = 0.6 * len(data)
        if y >= threshold:
            data.drop([x], 1, inplace=True)
        else:
            data[x] = data[x].fillna(data[x].mode()[0]) 
#             if data[x].dtypes == 'O':
#                 data[x] = data[x].fillna(data[x].mode()[0]) 
#             else:
#                 data[x] = data[x].fillna(data[x].mean()) 
    print('Done')
```

```python
def train_score(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train.ravel())
    pred_y = model.predict(x_test)
    return mean_squared_error(y_test, pred_y)
```
